# Minimal example

```python
# Describe a bullet and give its muzzle velocity:
mediumBullet = Bullet(0.22,  # Ballistic Coefficient (G7)
                      0.308, # Caliber (inches)
                      168,   # Weight (grains)
                      1.22,  # Length (inches)
                      2600)  # Muzzle velocity (fps)
# Set the zero distance:
zero = Distance(400, DistanceYard)

# Instantiate the calculator:
c = Calculator(mediumBullet)
# Here are some calculations it can do:
aim = c.elevationForZeroDistance(zero)  # This is the aiming angle to hit the zero with all default parameters
print(f'Created {c}'
      f'\nBarrel elevation for {zero}{UNIT_DISPLAY[c.distanceUnits].units} zero: {aim:.1f}{UNIT_DISPLAY[c.angleUnits].units}'
      f'\nDanger space for 10" target at zero distance: {c.dangerSpace(c.zeroGivenElevation(aim), 10):.0f}{UNIT_DISPLAY[c.distanceUnits].units}'
      )
```

```python
# Request a trajectory table in 100-yard increments out to 1000 yards
c.trajectory(1000, 100, elevation=aim)
```

## Plot

```python
import matplotlib.pyplot as plt
df = c.trajectory(1000, 10, zeroDistance=zero.value())  # Step in 10-yard increments to produce smoother curves
ax = df.plot(y=[c.tableCols['Drop'][0]], ylabel=UNIT_DISPLAY[c.heightUnits].units)
df.plot(y=[c.tableCols['Velocity'][0]], ylabel=UNIT_DISPLAY[c.bullet.velocityUnits].units, secondary_y=True, ylim=[0, df[c.tableCols['Velocity'][0]].max()], ax=ax)
plt.show()
```
